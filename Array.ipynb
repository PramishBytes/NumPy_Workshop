__ARRAY__

__To create a NumPy array, you cna use the function np.array()__





```python
import numpy as np 
a = np.array([1,2,3])
print(a)
print(a.ndim) # ndim provides the dimension of the array
```

```python
x= [1,2,3,4,5]

y = np.array(x)
print(y)

print(type(y))
```

```python
l = []

for i in range(1,5):
    int_1 = int(input("enter: "))
    l.append(int_1)

print(np.array(l))
```
