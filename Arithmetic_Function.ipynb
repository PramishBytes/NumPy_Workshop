__Arithmetic Functions__

1) np.min(x)
2) np.max(x)
3) np.argmin(x)
4) np.sqrt(x)
5) np.sin(x)
6) np.cos(x)
7) np.cumsum(x)

```python
import numpy as np 

var = np.array([1,2,3,4,5,2])

print("min:",np.min(var))
```

```python
print("max:",np.max(var))
```

```python
print("argmin: ",np.argmin(var)) #argument minimum
print("argmax: :" ,np.argmax(var))
```

```python
var1 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(np.min(var1))
```

```python
var1 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(np.max(var1))
```

```python
var1 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(np.sqrt(var1)) #squareroot
```

```python
var1 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(np.sin(var1))
```

```python
var1 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(np.cumsum(var1))  #cumsum calculates the cummulative sum of array elements. It returns a new array of the same shape as the input array,where each element at index i is the sum of all elements up to and including index i in the original array.b
```
