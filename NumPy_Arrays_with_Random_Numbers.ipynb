__Create NumPy Arrays with Random Numbers__

rand() : the function is used to generate a random vlaue between 0 to 1.

randn(): the function is used to generate a random value close to zero. This may return positive or negative numbers as well.

ranf(): the function for doing random sampling in numpy.It resturns an array of specified shape and fills it with random floats in the half-open inteval [0.0,1.0]

randint(): the function is used to generate a random integers within a specified range (inclusive of both endpoints)


__Random__


__Rand()__

```python
import numpy as np

var = np.random.rand(4)
print(var)
```

```python
print(var)
```

```python
var1 = np.random.rand(2,5)
print(var1)
```


__Randn()__

The fundtion is used to generate a random value close to zero.
This may return positive or negative numbers as well

```python
var2 = np.random.randn(4)
print(var2)
```

```python
print(var2)
```

```python
var2_1 = np.random.randn(3,4)
print(var2_1)
```

__ranf()__

the function for doing random sampling in numpy. It returns an array of specified shape and fills it with random floats in the half-open interval [0.0,1.0]


```python
var3 = np.random.ranf(5)
print(var3)
```

```python
print(var3)
```

```python

var33 = np.random.ranf((3,3,3))
print(var33)
```

__randint()__

the function is used to generate a random integers within a specified range(inclusive of both endpoints) 


```python
var4= np.random.randint(4)
print(var4)
```

```python
var44 = np.random.randint((2,4,6))
print(var44)
```

```python
print(var4)
```
