__Arithmetic Operation__

1) a+b -> np.add(a,b)
2) a-b -> np.subtract(a,b)
3) a*b -> np.multiply(a,b)
4) a/b -> np.divide(a,b)
5) a%b -> np.mod(a,b)
6) a**b -> np.power(a,b)
7) 1/a -> np.reciprocal(a)

```python
import numpy as np 

var = np.array([1,2,3,4])

varadd = var + 3
print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1+var2
print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1/var2
print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1%var2
print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1**var2
print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1-var2
print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = 1/var2
print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1*var2
print(varadd)
```
