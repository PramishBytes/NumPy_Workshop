__Data Types In NumPy Arrays__

```python
import numpy as np 

var = np.array([1,2,3,4])
print("Data Type: ",var.dtype)
```

```python
var = np.array([1.2,2.2])
print("Data Type: ",var.dtype)
```

```python
var = np.array(['a','b','b',1,2,3,4])
print("Data Type: ", var.dtype)  #Unicode 
```

```python
x= np.array([1,2,3],dtype = np.int8)
print("Data Type: ",x.dtype)
print(x)
```

```python
x= np.array([1,2,3],dtype = "f")
print("Data Type: ",x.dtype)
print(x)
```

```python
x2 = np.array([1,2,3,4,5])
new = np.float32(x2)
print("Data Type: ",x2.dtype)
print("Data Type: ",new.dtype)

print(x2)
print(new)

new_one = np.int_(new)
print(new_one)
```

```python
x3 = np.array([1,2,3,4])

new_1 = x3.astype(float)

print(x3)
print(new_1)
```
