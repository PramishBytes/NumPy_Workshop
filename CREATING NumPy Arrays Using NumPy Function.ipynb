CREATING NumPy Arrays Using NumPy Function

Special NumPy Array

Array filled with 0's
Array filled with 1's
Create an empty array
An array with a range of elements
Array diagonal element filled with 1's
Create an array with values that are spaced linearly in a specified interval

```python
import numpy as np 

ar_zero = np.zeros(4)
ar_zero1 = np.zeros((3,4))
print(ar_zero)
print()
print(ar_zero1)
```

__Ones__


```python
ar_one = np.ones(4)

print(ar_one)
```

__Empty__


```python
ar_em = np.empty(4)
print(ar_em) #Provides the same output as above is due to the memory recalling it 
```

__Range__

```python
ar_rn = np.arange(4)
print(ar_rn)
```

__Diagonal__

```python
ar_dia = np.eye(3)
print(ar_dia)
```

```python
ar_dia = np.eye(5,5)
print(ar_dia)
```

__linspace function__

generates an array of evenly spaced numbers over a difined interval

```python
ar_lin = np.linspace(1,10,num=5)
print(ar_lin)
```
